```python
# 1.
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv"
df = pd.read_csv(url)

# Filter data based on gender
male_survivors = df[df['Sex'] == 'male']
female_survivors = df[df['Sex'] == 'female']

# Plot Male survivors
plt.scatter(male_survivors['Age'], male_survivors['Fare'], color='green', label='Male')

# Plot Female survivors
plt.scatter(female_survivors['Age'], female_survivors['Fare'], color='red', label='Female')

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatter Plot of Male and Female Survivors")

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
# 2.
# Filter data for passengers who did not survive
male_dead = df[(df['Sex'] == 'male') & (df['Survived'] == 0)]
female_dead = df[(df['Sex'] == 'female') & (df['Survived'] == 0)]

# Plot Male dead passengers
plt.scatter(male_dead['Age'], male_dead['Fare'], color='blue', label='Male')

# Plot Female dead passengers
plt.scatter(female_dead['Age'], female_dead['Fare'], color='yellow', label='Female')

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatter Plot of Male and Female Dead Passengers")

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
# 3.
# Create a new column 'Number of Relatives' as the sum of SibSp and Parch
df['Number of Relatives'] = df['SibSp'] + df['Parch']

# Display the first few rows to verify
print(df[['SibSp', 'Parch', 'Number of Relatives']].head())

# Filter data for passengers who did not survive
dead_passengers = df[df['Survived'] == 0]

# Plot Dead passengers with 'Number of Relatives' on the X-axis and 'Fare' on the Y-axis
plt.scatter(dead_passengers['Number of Relatives'], dead_passengers['Fare'], color='#4682B4')  # SteelBlue color

# Add labels and title
plt.xlabel("Number of Relatives")
plt.ylabel("Fare")
plt.title("Scatter Plot of Dead Passengers - Number of Relatives vs Fare")

# Show the plot
plt.show()

```

```python
# 4.

# Create a new column 'Number of Relatives' as the sum of SibSp and Parch
df['Number of Relatives'] = df['SibSp'] + df['Parch']

# Filter data for passengers who were alive
alive_passengers = df[df['Survived'] == 1]

# Plot alive passengers with 'Number of Relatives' on the X-axis and 'Fare' on the Y-axis
plt.scatter(alive_passengers['Number of Relatives'], alive_passengers['Fare'], color='#4169E1')  # Royal Blue color

# Add labels and title
plt.xlabel("Number of Relatives")
plt.ylabel("Fare")
plt.title("Scatter Plot of Alive Passengers - Number of Relatives vs Fare")

# Show the plot
plt.show()
```
